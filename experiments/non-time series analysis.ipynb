```python
import warnings
warnings.simplefilter('ignore')

import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv('../datasets/processed_dataset.csv')
df
```

```python
df.drop([df.columns[0], df.columns[1], df.columns[2]], inplace=True, axis=1)   # dropping course details
```

```python
df = pd.concat([df, pd.get_dummies(df['Course Year'], prefix='Course Year'), pd.get_dummies(df['Department Code'], prefix='Department Code'), pd.get_dummies(df['Course Level'], prefix='Course Level'), pd.get_dummies(df['Standing'], prefix='Standing'), pd.get_dummies(df['Status'], prefix='Status')], axis=1)
df.drop(['Course Year', 'Department Code', 'Course Level', 'Status', 'Standing'], axis=1, inplace=True)
```

```python
le = LabelEncoder()
le.fit(['A+', 'A', 'A-', 'B+', 'B', 'B-', 'C+', 'C', 'C-', 'D+', 'D', 'D-', 'F'])
```

```python
def get_error_score(df, model):    
    X = df.drop('Semester', axis=1)
    y = le.transform(X.pop('Letter Grade'))
    col = X.columns
    
    sc = StandardScaler()
    X = pd.DataFrame(sc.fit_transform(X), columns=col)
    
    error_scores = {'mae': [], 'rmse': []}
    skf = StratifiedKFold(n_splits=7)
    X = X.to_numpy()
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
    
        model.fit(X_train, y_train)
        
        y_pred = model.predict(X_test)
        rmse = round(np.sqrt(mean_squared_error(y_test, y_pred)), 2)
        mae = round(mean_absolute_error(y_test, y_pred), 2)
        error_scores['rmse'].append(rmse)
        error_scores['mae'].append(mae)
        
    return error_scores
```

```python
error_scores = get_error_score(df, GradientBoostingRegressor(learning_rate=0.01, 
                                                             loss='huber',
                                                             n_estimators=800,
                                                             max_depth=3))
```

```python
with open('../results/baseline_regression_results.json') as fr:
    reg_results = json.load(fr)
    best_score = reg_results['GradientBoostingRegressor']
```

```python
def plot_error_scores(non_time_series, best_score):
    
    best_model_scores = {'mae': [], 'rmse': []}
    for n_training_sem in best_score:
        rmse = best_score[n_training_sem]['rmse'][1]
        mae = best_score[n_training_sem]['mae'][1]
        best_model_scores['rmse'].append(rmse)
        best_model_scores['mae'].append(mae)

    x = np.arange(7)
    fig, ax = plt.subplots(2, figsize=(7.8, 5), sharex=True)

    ax[0].bar(x - 0.225, non_time_series['rmse'], 0.45, label='Cross Validation', hatch='xx', color='white')
    ax[0].bar(x + 0.225, best_model_scores['rmse'], 0.45, label='Time Series', hatch='||', color='white')
    ax[1].bar(x - 0.225, non_time_series['mae'], 0.45, label='Cross Validation', hatch='xx', color='white')
    ax[1].bar(x + 0.225, best_model_scores['mae'], 0.45, label='Time Series', hatch='||', color='white')

    for i, v in enumerate(non_time_series['rmse']):
        ax[0].text(i-0.45, v + 0.1, '{:.2f}'.format(round(v, 2)), fontfamily='serif', fontsize=10, fontweight='bold')
        
    for i, v in enumerate(best_model_scores['rmse']):
        ax[0].text(i+0.02, v + 0.1, '{:.2f}'.format(round(v, 2)), fontfamily='serif', fontsize=10, fontweight='bold')        

    for i, v in enumerate(non_time_series['mae']):
        ax[1].text(i-0.45, v + 0.1, '{:.2f}'.format(round(v, 2)), fontfamily='serif', fontsize=10, fontweight='bold')
        
    for i, v in enumerate(best_model_scores['mae']):
        ax[1].text(i+0.02, v + 0.1, '{:.2f}'.format(round(v, 2)), fontfamily='serif', fontsize=10, fontweight='bold')        
        
    ax[0].set_ylabel('RMSE', fontsize=11, fontfamily='serif')
    ax[1].set_ylabel('MAE', fontsize=11, fontfamily='serif')
    ax[1].set_xlabel('Folds', fontsize=11, fontfamily='serif')
    
    ax[0].set_yticks([0.0, 0.5, 1.0, 1.5, 2.0, 2.5])
    ax[1].set_yticks([0.0, 0.5, 1.0, 1.5, 2.0])
    ax[0].set_yticklabels(ax[0].get_yticks(), fontsize=11, fontfamily='serif')
    ax[1].set_yticklabels(ax[1].get_yticks(), fontsize=11, fontfamily='serif')
    ax[1].set_xticks(x)
    ax[1].set_xticklabels([i+1 for i in range(7)], fontsize=11, fontfamily='serif')

    ax[0].grid(axis='y')
    ax[1].grid(axis='y')
    
    ax[0].spines["right"].set_visible(False)
    ax[0].spines["left"].set_visible(False)
    ax[0].spines["top"].set_visible(False)
    ax[1].spines["right"].set_visible(False)
    ax[1].spines["left"].set_visible(False)
    ax[1].spines["top"].set_visible(False)
    
    ax[0].tick_params(axis = "x", which = "both", bottom = False, top = False)
    ax[0].tick_params(axis = "y", which = "both", left = False, right = False)    
    ax[1].tick_params(axis = "x", which = "both", bottom = False, top = False)
    ax[1].tick_params(axis = "y", which = "both", left = False, right = False)
    
    ax[1].legend(prop={'size': 13, 'family': 'serif'}, loc='center', bbox_to_anchor=(0.5, -0.36), ncol=4)
    plt.savefig('non-time series comparison.eps', bbox_inches='tight', dpi=1200)
```

```python
plot_error_scores(error_scores, best_score)
```
