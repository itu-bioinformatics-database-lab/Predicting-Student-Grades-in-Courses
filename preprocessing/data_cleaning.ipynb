```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/original_dataset.csv')
```

```python
df
```

```python
df.rename(columns={'Kod': 'Course Code', 
                   'Tanım': 'Course Title', 
                   'Öğrenci No': 'Student Number', 
                   'Bölüm': 'Department Code', 
                   'Bölüm (tanım)': 'Course Level', 
                   'Not': 'Letter Grade', 
                   'Kayıt drm.': 'Status', 
                   'Dönem': 'Semester'},
         inplace=True)
df.drop('Skala', axis=1, inplace=True)
df.head()
```

### Updating Course Level
In Istanbul Sehir University, courses are numbered after their level. Therefore, we use the starting number of course code to determine their level. The following updates the course level for each course.

```python
for row_idx in df.index:
    course_code = df.iloc[row_idx, 0]
    if course_code.split()[1][0] in ['1', '2', '3', '4']:
        df.iloc[row_idx, 4] = 'Undergraduate'
    elif course_code.split()[1][0] in ['5', '6']:
        df.iloc[row_idx, 4] = 'Graduate'
    else:
        df.iloc[row_idx, 4] = 'PhD'
```

### Updating semester names to English

```python
for row_idx in df.index:
    semester_name = df.iloc[row_idx, 7].split(' - ')
    if semester_name[1] == 'Güz':
        df.iloc[row_idx, 7] = semester_name[0] + ' - ' + 'Fall'
    elif semester_name[1] == 'Bahar':
        df.iloc[row_idx, 7] = semester_name[0] + ' - ' + 'Spring'
    else:
        df.iloc[row_idx, 7] = semester_name[0] + ' - ' + 'Summer'
```

### Updating Status

```python
for row_idx in df.index:
    if df.iloc[row_idx, 6] == 'Başarıyla tamamlandı':
        df.iloc[row_idx, 6] = 'Successful'
    elif df.iloc[row_idx, 6] == 'Başarısız tamamlandı':
        df.iloc[row_idx, 6] = 'Unsuccessful'
    else:
        df.iloc[row_idx, 6] = 'Recorded'
```

### Loading up-to-date course information
The most up-to-date course information has been scraped using selenium from https://www.sehir.edu.tr/en/academics/all-programs

```python
course_df = pd.read_csv('../datasets/updated_course_info.csv')
course_df
```

### Updating outliers information
Some courses have been changed completely in terms of their course code and course name. Course name is being when deciding to change course information (i.e. course code, course year, ...). All courses whose information have been completely changed might not be listed here.

```python
for row_idx in df.index:
    course_code = df.iloc[row_idx, 0]
    if course_code == 'UNI 113':
        df.iloc[row_idx, 1] = 'Textual Analysis and Effective Communication'
    elif course_code == 'UNI 114':
        df.iloc[row_idx, 1] = 'Textual Analysis and Academic Writing'
```

### Adding new feature columns to the main dataframe

```python
df.insert(8, 'Theoritical', 3)
df.insert(9, 'Practical', 0)
df.insert(10, 'Course Credit', 3)
df.insert(11, 'ECTS', 5)
df.insert(12, 'Course Year', 1)
```

### Updating old course details

```python
for row_idx in df.index:
    course_code = df.iloc[row_idx, 0]
    course_title = df.iloc[row_idx, 1]
    title_indices = list(course_df.iloc[:, 1][course_df.iloc[:, 1] == course_title].index)
    code_indices = course_df.iloc[:, 0][course_df.iloc[:, 0] == course_code].index
    for i in title_indices:
        cc_1 = course_code.split()   # old course code
        cc_2 = course_df.iloc[i, 0].split()   # new course code
        if course_title == course_df.iloc[i, 1] and 'Summer' not in course_title and cc_1[0] == cc_2[0] and cc_1[1] != cc_2[1]:
            df.iloc[row_idx, 0] = course_df.iloc[i, 0]
            df.iloc[row_idx, 8] = course_df.iloc[i, 4]
            df.iloc[row_idx, 9] = course_df.iloc[i, 5]
            df.iloc[row_idx, 10] = course_df.iloc[i, 2]
            df.iloc[row_idx, 11] = course_df.iloc[i, 3]
```

### Updating course year based on the first integer of course code

```python
for idx in df.index:
    df.iloc[idx, 12] = df.iloc[idx, 0].split()[1][0]
```

### Dropping instances with a grading scale other than A, B, C, D, F

```python
unweight_grades = []
for row_idx in df.index:
    if df.iloc[row_idx, 5] in ['IA', 'S', 'U', 'NP', 'TP']:
        unweight_grades.append(row_idx)        
for idx in unweight_grades:
    df.drop(idx, inplace=True)
print(len(unweight_grades), 'rows will be removed because they have unweighted grades')
df.index = range(len(df))
```

### Dropping instances from students with less than 5 courses 

```python
temp = df[['Student Number', 'Letter Grade']].groupby(['Student Number'], as_index=False).count().sort_values(by=['Letter Grade'])
temp.index = range(len(temp))
c = 0
unqualified_stds = []
for i in temp.index:
    student_number = temp.iloc[i, 0]
    total_completed = temp.iloc[i, 1]
    if total_completed < 5:
        c += total_completed
        unqualified_stds.append(student_number)
        
for student_number in unqualified_stds:
    df.drop(df[df['Student Number']==student_number].index, inplace=True)

df.index = range(len(df))
print(c, ' rows will be removed when we want to eliminate students who have completed less than 5 courses.')
```

```python
df.to_csv('cleaned_dataset.csv', index=False)
```
